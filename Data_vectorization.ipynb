```python
df['id']=df['id'].astype('str')
```

```python
df = df.merge(df[['id', 'definition']], left_on='parent', right_on='id', how='left', suffixes=('', '_parent1'))
df.rename(columns={'definition_parent1': 'parent_description1'}, inplace=True)
```

```python
df_vectorization = df.copy()

# Define the columns to include in the text
text_columns = [
    'title', 'definition', 'full_text',
     'exclusions', 'parent_description1'
]

# Function to safely build the combined text field
def build_vectorization_text(row):
    parts = []
    for col in text_columns:
        if pd.notna(row.get(col, "")) and str(row[col]).strip() != "":
            label = col.replace('_', ' ').capitalize()
            parts.append(f"{label}: {str(row[col]).strip()}")
    return "\n".join(parts)

# Apply the function to build the new column
df_vectorization['vectorization_text'] = df_vectorization.apply(build_vectorization_text, axis=1)
```
